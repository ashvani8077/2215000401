```python
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-5,5,200)
```

```python
def binary_step(x):
    return np.where(x >= 0, 1, 0)
plt.plot(x, binary_step(x))
plt.title("Binary Step Activation Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
