```python
import math

def sigmoid(x):
  return 1/(1+math.exp(-x))
```

```python
sigmoid(1)
```

```python
sigmoid(-56)
```

```python
sigmoid(0.5)
```

```python
def tanh(x):
  return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))
```

```python
tanh(-56)
```

```python
tanh(50)
```

```python
tanh(1)
```

```python
def relu(x):
  return max(0,x)
```

```python
relu(-7)
```

```python
relu(2)
```

```python
def leaky_relu(x):
  return max(0.1*x,x)
```

```python
leaky_relu(-5)
```

```python
leaky_relu(5)
```
