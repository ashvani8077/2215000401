```python
inputs=[
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]
expected_outputs=[0,1,1,0]
```

```python
def activation_function(value):
  if value>0:
    return 1
  else:
    return 0
```

```python
def OR_function(x1,x2):
  w1=1
  w2=1
  b=-0.5
  net_input=x1*w1+x2*w2+b
  return activation_function(net_input)
```

```python
def AND_function(x1,x2):
  w1=1
  w2=1
  b=-1.5
  return activation_function(x1*w1+x2*w2+b)
```

```python
def NAND_function(x1,x2):
  w1=-1
  w2=-1
  b=1.5
  return activation_function(x1*w1+x2*w2+b)
```

```python
def XOR_function(x1, x2):
  h1=OR_function(x1,x2)
  h2=NAND_function(x1,x2)
  output=AND_function(h1,h2)
  return output
```

```python
print("XOR Gate table using Perceptron \n")
for i in range(len(inputs)):
  x1,x2=inputs[i]
  output=XOR_function(x1,x2)
  print(f"Input: {x1},{x2} -> Output: {output}")
```
