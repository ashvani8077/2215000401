```python
inputs=[
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]
expected_output=[0,0,0,1]
```

```python
w1=1
w2=1
b=-1.5
```

```python
def activation_function(value):
  if value>0:
    return 1
  else:
    return 0
```

```python
print("AND gate table using Perceptron \n")
for i in range(len(inputs)):
  x1,x2=inputs[i]
  net_input=(x1*w1)+(x2*w2)+b
  output=activation_function(net_input)
  print(f"Input: {x1},{x2} -> Output: {output}")
```
