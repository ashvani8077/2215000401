```python
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-5,5,200)
```

```python
def ramp(x):
    return np.where(x < 0, 0, np.where(x <= 1, x, 1))
plt.plot(x, ramp(x))
plt.title("Ramp Activation Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
