```python
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-5,5,200)
```

```python
def tanh(x):
  return np.tanh(x)
plt.plot(x, tanh(x))

plt.title("Tanh Activation Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
