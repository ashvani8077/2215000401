```python
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-5,5,200)
```

```python
def logistic_sigmoid(x):
  return 1 / (1+np.exp(-x))
plt.plot(x, logistic_sigmoid(x))
plt.title("Logistic Sigmoid Activation Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
