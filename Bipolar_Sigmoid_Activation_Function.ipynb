```python
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-5,5,200)
```

```python
def bipolar_sigmoid(x):
  return (2/ (1 + np.exp(-x))) - 1
plt.plot(x, bipolar_sigmoid(x))
plt.title("Bipolar Sigmoid Activation Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
